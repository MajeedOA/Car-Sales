```python
pd.set_option("display.max_rows", 237)
```

```python

df["average_monthly_sales"] =(df["Sumofsales"]/12).astype(int)
df["Q1 Sales"] = df[["Jan", "Feb", "Mar"]].sum(axis=1)
df["Q2 Sales"] = df[["Apr", "May", "Jun"]].sum(axis=1)
df["Q3 Sales"] = df[["Jul", "Aug", "Sep"]].sum(axis=1)
df["Q4 Sales"] = df[["Oct", "Nov", "Dec"]].sum(axis=1)
df["H1_Sales"] = df.iloc[:, 1:7].sum(axis=1)  # Jan to Jun sales
df["H2_Sales"] = df.iloc[:, 7:13].sum(axis=1) # Jul to Dec sales
df["Peak_Month"] = df.iloc[:, 1:13].idxmax(axis=1) # Month with highest sales
df
```


 1. Find the highest and lowest selling car models

```python

highest_selling_model = df.loc[df["Sumofsales"].idxmax(), ["Model", "Sumofsales"]]
lowest_selling_model = df.loc[df["Sumofsales"].idxmin(), ["Model", "Sumofsales"]]

lowest_selling_model,highest_selling_model

# Plot a bar chart
plt.figure(figsize=(16, 10))
plt.bar(df["Model"],df["Sumofsales"], color=["green", "red"])
plt.xlabel("Car Model")
plt.ylabel("Total Sales")
plt.title("Highest and Lowest Selling Car Models")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()









```

2. Calculate H1 vs H2 sales comparison

```python

h1_h2_sales_comparison = df[["H1_Sales", "H2_Sales"]].sum()
h1_h2_sales_comparison

# Plot a bar chart
plt.figure(figsize=(8, 6))
h1_h2_sales_comparison.plot(kind='bar', color=['blue', 'orange'])
plt.xlabel("Sales Period")
plt.ylabel("Total Sales")
plt.title("H1 vs H2 Sales Comparison")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()



```

3. Find the most common peak sales month

```python

most_common_peak_month = df["Peak_Month"].mode()[0]

most_common_peak_month

# Count the occurrences of each peak month
peak_month_counts = df["Peak_Month"].value_counts()

# Plot a bar chart
plt.figure(figsize=(10, 6))
peak_month_counts.plot(kind='bar', color='skyblue')
plt.xlabel("Peak Month")
plt.ylabel("Count")
plt.title("Most Common Peak Sales Month")
plt.xticks(rotation=0)
plt.axvline(x=most_common_peak_month, color='red', linestyle='--', label='Most Common Peak Month')
plt.legend()
plt.tight_layout()
plt.show()






```
