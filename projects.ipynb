```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd


df= pd.read_csv(r"C:\Users\MY LAPTOP\Downloads\Canadasalesdata.csv")
df
```

```python
pd.set_option("display.max_rows", 237)
```

```python
df["average_monthly_sales"] =df["Sumofsales"]/12
df
```

```python
df["average_monthly_sales"] =(df["Sumofsales"]/12).astype(int)
df
```

```python
df["Q1 Sales"] = df[["Jan", "Feb", "Mar"]].sum(axis=1)
df["Q2 Sales"] = df[["Apr", "May", "Jun"]].sum(axis=1)
df["Q3 Sales"] = df[["Jul", "Aug", "Sep"]].sum(axis=1)
df["Q4 Sales"] = df[["Oct", "Nov", "Dec"]].sum(axis=1)
df
```

```python
Model = df.groupby('Model')['Sumofsales'].sum()
Model
```

```python
total_sales = df['Sumofsales'].sum()
total_sales
```

```python
df['Market Share(%)'] =(df['Sumofsales']/total_sales) * 100
df
```

```python
df['Sumofsales'].mean()
```

```python
Sales_range = df['Sumofsales'].max() - df['Sumofsales'].min()
Sales_range
```

```python
# Calculate total sales for each year
total_sales_2019 = df[df["Year"] == 2019]["Sumofsales"].sum()
total_sales_2020 = df[df["Year"] == 2020]["Sumofsales"].sum()
total_sales_2021 = df[df["Year"] == 2021]["Sumofsales"].sum()

print("Total Sales in 2019:", total_sales_2019)
print("Total Sales in 2020:", total_sales_2020)
print("Total Sales in 2021:", total_sales_2021)
```

```python
df["H1_Sales"] = df.iloc[:, 1:7].sum(axis=1)  # Jan to Jun sales
df["H2_Sales"] = df.iloc[:, 7:13].sum(axis=1) # Jul to Dec sales
df["Peak_Month"] = df.iloc[:, 1:13].idxmax(axis=1) # Month with highest sales
df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Category", hue='Category', y="Sumofsales", data=df, estimator=sum, errorbar=None, palette="muted")
plt.title("Sum of Sales by Category")
plt.ylabel("Sum of Sales")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Year', y='Sumofsales', hue='Year', palette='Set2')
plt.xlabel('Year')
plt.ylabel('Sum of Sales')
plt.title('Sum of Sales by Year')
plt.show()
```

```python
monthly_avg = df.iloc[:, 1:13].mean()
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_avg.index, y=monthly_avg.values, marker="o", color="b")
plt.title("Average Monthly Sales")
plt.ylabel("Sum of Sales")
plt.xlabel("Month")
plt.show()

```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x=df["Jan"], y=df["Dec"], alpha=0.6)
plt.title("Jan vs Dec Sales")
plt.xlabel("Jan Sales")
plt.ylabel("Dec Sales")
plt.show()
```

```python
# Plot histograms for numeric columns 
fig, ax = plt.subplots(2, 2, figsize=(12, 10))  # 2 row, 2 columns

# Histogram for total sales
ax[0, 0].hist(df["Sumofsales"], bins=20, edgecolor='blue')
ax[0, 0].set_title("Distribution of Total Sales")

# Histogram for H1 Sales
ax[0, 1].hist(df["H1_Sales"], bins=6, color='red', edgecolor='black')
ax[0, 1].set_title("Distribution of H1 (Jan-Jun) Sales")

# Histogram for H2 Sales
ax[1, 0].hist(df["H2_Sales"], bins=6, color='green', edgecolor='white')
ax[1, 0].set_title("Distribution of H2 (Jul-Dec) Sales")

# Histogram for Average Monthly Sales
ax[1, 1].hist(df["average_monthly_sales"], bins=20, color='purple', edgecolor='white')
ax[1, 1].set_title("Distribution of Average Monthly Sales")



```

```python
# 1. Find the highest and lowest selling car models
highest_selling_model = df.loc[df["Sumofsales"].idxmax(), ["Model", "Sumofsales"]]
lowest_selling_model = df.loc[df["Sumofsales"].idxmin(), ["Model", "Sumofsales"]]

lowest_selling_model,highest_selling_model



```

```python
# 2. Compare H1 vs H2 Sales
h1_h2_sales_comparison = df[["H1_Sales", "H2_Sales"]].sum()

h1_h2_sales_comparison

```

```python
# 3. Find the most common peak sales month
most_common_peak_month = df["Peak_Month"].mode()[0]

most_common_peak_month

```

```python

# 4. Compare Peak month with Sum of Sales
plt.figure(figsize=(6,4))
sns.barplot(x="Peak_Month", y="Sumofsales", hue='Peak_Month', data=df, palette="pastel")
plt.title("Sales Distribution by Peak Month")
plt.ylabel("Sum of Sales")
plt.xlabel("Peak Month")
plt.xticks(rotation=45)
```

```python
# Compute summary statistics for numerical columns
summary_stats = df.describe().T

# Compute range (max - min) for each numeric column
summary_stats["range"] = summary_stats["max"] - summary_stats["min"]

summary_stats
```

```python
# Summarizing total sales per year
sales_per_year = df.groupby("Year")["Sumofsales"].sum()
sales_per_year
```

```python

# Plot sales trend over the years
plt.figure(figsize=(10, 5))
sns.lineplot(x=sales_per_year.index, y=sales_per_year.values, marker="o", linestyle="-", color="b")
plt.title("Total Car Sales in Canada Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)
```

```python
# Summarizing total sales per category
sales_per_category = df.groupby("Category")["Sumofsales"].sum()
sales_per_category
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Category", hue='Category', y="Sumofsales", data=df, estimator=sum, errorbar=None, palette="muted")
plt.title("Sum of Sales by Category")
plt.ylabel("Sum of Sales")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 best-selling car models
top_models = df.nlargest(10, "Sumofsales")
top_models
```

```python
# Plot top 10 car models by sales
plt.figure(figsize=(12, 6))
sns.barplot(x=top_models["Sumofsales"], y=top_models["Model"], hue=top_models["Model"], palette="viridis")
plt.title("Top 10 Best-Selling Car Models in Canada")
plt.xlabel("Total Sales")
plt.ylabel("Car Model")
plt.show()
```

Conclusion

After analyzing the car sales data in Canada, there was a noticeable drop in total sales from 2019 to 2021, which could be due to external factors like economic shifts or supply chain disruptions. Mid-size SUVs (MS category) had the highest sales, showing that consumers prefer larger vehicles, possibly for their practicality and comfort. Among individual models, the Jeep Wrangler was one of the best-selling vehicles, reinforcing the popularity of SUVs in the market.

This analysis provides useful insights,graphs were plotted to visualize the sales trends to show differences in the column. It would be interesting to compare these trends with electric vehicle (EV) sales or look at how pricing impacts consumer choices. Additionally, examining regional trends across Canada could give a better picture of where demand is strongest.

Overall, this dataset gave a solid overview of sales trends, and with more data, we could dive even deeper into what drives the Canadian automotive market.
